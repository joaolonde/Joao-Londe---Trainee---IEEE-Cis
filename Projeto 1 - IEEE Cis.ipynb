```python
class Circulo():
  pi = 3.14                          #Defini global
  def __init__ (self, raio = 2):
    self.raio = raio

  
  def area(self):
    return (self.raio * self.raio) * Circulo.pi   #para usar as variáveis globais aqui dentro precisa ser assim

  def setRaio(self, novoraio):
    self.raio = novoraio

  def getRaio(self):
    return self.raio        
```

```python
#Formas de operar com o circulo

circ = Circulo()
circ.getRaio()
circ.area()
circ.setRaio(8)
circ.getRaio()

circ = Circulo(9)
circ.getRaio()
circ.area()
```

```python
df1 = pd.read_excel("Acoes_fundamentus.xlsx")
df2 = pd.read_excel("Ibovespa.xlsx")

acoes = df1
bov = df2

#Renomeia as colunas do 1° dataset e remove dados faltantes
acoes.columns
acoes = acoes.rename(columns = {'Patrim. Líq': 'Patrimonio Liquido', 'Liq.2meses': 'Volume medio de negocioções (2M)', 'Liq. Corr.': 'Liquidez corrente', 'Cresc. Rec.5a':'Crescimento da receita (5A)'}, inplace = False)
acoes.columns
acoes.dropna()


#Renomeia as colunas do 2° dataset e remove dados faltantes
bov.columns
bov = bov.rename(columns = {'Código':'Papel', 'Part. (%)':'Participação (%)'}, inplace = False)
bov.dropna()

#Remove discrepancias
acoes = acoes.drop(acoes[acoes['P/L'] < -200].index)
acoes = acoes.drop(acoes[acoes['P/L'] > 200].index)
acoes = acoes.drop(acoes[acoes['Div.Yield'] > 0.3].index)
acoes = acoes.drop(acoes[acoes['Mrg Ebit'] > 1].index)
acoes = acoes.drop(acoes[acoes['Mrg Ebit'] < -1].index)



qtd_acoes = acoes.shape[0]
qtd_bov = bov.shape[0]

print(qtd_acoes, qtd_bov)
acoes.sort_values(by=['Div.Yield'], ascending=False)
```

# **Análise 2**

Agora, queremos relacionar o a Margem Ebit de uma empresa com o seu Dividend Yield. A análise foi feita no dataset que contém todas as ações da bolsa.

Na prática, o Ebit de uma empresa representa o lucro operacional da empresa, antes de descontar impostos e juros. A margem Ebit representa a divisão do lucro operacional pela receita líquida, na prática, o que sobrou após pagar os custos dos produtos as despesas com vendas.

Já o Dividend Yeld, está relacionado a distribuição de lucros da empresa para os acionistas. O índice é uma porcentagem, e essa porcentagem indica a fatia do valor da ação da empresa, que o detentor daquela ação recebeu no acumulado do ano anterior.

A ideia é relacionar os indicadores de forma a identificar se existe relação entre a Margem Ebit e o Yield da empresa, mostrando assim se empresas que tem uma fração de lucro maior, pagam dividendos maiores.

Para a análise, serão consideradas apenas empresas com Margem Ebit Positiva. Ou seja, empresas que tiveram lucro.

```python
acoes.columns
#Remove as Margens Ebit negativas
lucro = acoes.drop(acoes[acoes['Mrg Ebit'] < 0].index)
lucro = lucro.drop(lucro[lucro['Div.Yield'] == 0].index)

#Correlação dos dados
corr = lucro['Div.Yield'].corr(lucro['Mrg Ebit'])
print('Correlação: ', corr)

#Variáveis com as colunas
yield_ = lucro['Div.Yield']
ebit = lucro['Mrg Ebit']

print('Yield médio: ', yield_.mean())
print("Margem Ebit Média: ", ebit.mean())

#Gráfico gerado automaticamente
graf = sns.lmplot('Div.Yield', 'Mrg Ebit', lucro).set(title='yield x ebit - Gerado automaticamente')


graf = sns.lmplot('Div.Yield', 'Mrg Ebit', lucro).set(title='yield x ebit - Dados relevantes')
graf.set(ylim=(0, 1)) # 1 equivale a 100%
graf.set(xlim=(0, 0.3)) # 0,3 equivale a 30%

```

Para este caso, não podemos afirmar que empresas com maior geração de lucro pagam dividendos maiores, pois a análise mostra que a correlação é fraca. 

Uma possível explicação para este fenomeno, é o fato das empresas não serem obrigadas a distribuir dividendos, fazendo assim com que distribuam apenas aquilo que julgam adequado para sua situação financeira.
